#Dataset

```python
data = pd.read_csv("/content/drugs.csv")
```

# Data preprocessing

```python
df = data.copy()
```

```python
df.head()
```

```python
df.head()
```

```python
df.info()
```

```python
encoaded = ce.OrdinalEncoder(cols=[feature for feature in df.columns if df[feature].dtype == "O"]).fit_transform(df)
encoaded.head()
```

```python
encoaded.Cholesterol.unique()
```

```python
encoaded.isnull().sum()
```

# independent and dependent feature split

```python
# independent feature
x = encoaded.drop("Drug",axis=1)
# dependent feature
y = encoaded["Drug"]
```

# train test split

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

# randomforestclassifier

```python
model = RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

# model evaluation

```python
accuracy = accuracy_score(y_test,model.predict(x_test))
clasif_report = classification_report(y_test,model.predict(x_test))
confusion_mat = confusion_matrix(y_test,model.predict(x_test))
```

```python
print("accuracy :-", accuracy)
print("clasif_report :-", clasif_report)
print("confusion_mat :-", confusion_mat)
```
