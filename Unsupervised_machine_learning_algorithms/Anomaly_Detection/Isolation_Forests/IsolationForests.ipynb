```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.ensemble import IsolationForest
```

```python
dataset = pd.read_csv("/content/data_10.csv")
```

```python
dataset.head()
```

# data preprocessing

```python
data = dataset[["y","z"]]
```

```python
data.head()
```

```python
sns.scatterplot(x=data["y"], y=data["z"])
```

# isolation forests


```python
isf = IsolationForest(contamination="auto")
```

```python
isf.fit(data)
```

```python
pred = isf.predict(data)
```

```python
pred
```

```python
index = np.where(pred == -1)[0]
```

```python
index
```

```python
x = data.values
```

```python
sns.scatterplot(x=x[:,0], y=x[:,1])
sns.scatterplot(x=x[index,0], y=x[index,1], color="red")
```
